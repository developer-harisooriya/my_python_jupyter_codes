```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
categories = ['Category 1', 'Category 2', 'Category 3', 'Category 4', 'Category 5']
values1 = [25, 40, 30, 10, 35]
values2 = [15, 20, 10, 30, 25]
values3 = [10, 15, 20, 15, 10]

# Calculate the total sum for each category
total_sum = np.sum([values1, values2, values3], axis=0)

# Calculate the proportion for each value
proportion1 = np.divide(values1, total_sum) * 100
proportion2 = np.divide(values2, total_sum) * 100
proportion3 = np.divide(values3, total_sum) * 100

# Create a figure and axis
fig, ax = plt.subplots()

# Create stacked bars
ax.bar(categories, proportion1, label='Value 1')
ax.bar(categories, proportion2, bottom=proportion1, label='Value 2')
ax.bar(categories, proportion3, bottom=proportion1 + proportion2, label='Value 3')

# Set labels and title
plt.xlabel('Categories')
plt.ylabel('Percentage')
plt.title('100% Stacked Bar Graph Example')

# Add a legend
plt.legend()

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
categories = ['Category 1', 'Category 2', 'Category 3', 'Category 4', 'Category 5']
values = [25, 40, 30, 10, 35]

# Create a pie chart
plt.pie(values, labels=categories, autopct='%1.1f%%')

# Set title
plt.title('Pie Chart Example')

# Display the chart
plt.show()
```

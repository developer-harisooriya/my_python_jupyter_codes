```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 4, 5, 6, 7]
y = [2, 4, 6, 27, 10, 30]

# Create a figure and axis
fig, ax = plt.subplots()

# Plot the data as a line graph
ax.plot(x, y)

# Set labels and title
ax.set_xlabel('time')
ax.set_ylabel('distance')
ax.set_title('data of a vehicle')

# Display the graph
plt.show()
```
